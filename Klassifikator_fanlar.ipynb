Logistic Regression

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Convert to long format
df = df_final.melt(var_name="category", value_name="text").dropna().reset_index(drop=True)

# Balance categories
min_samples = df["category"].value_counts().min()
df_balanced = df.groupby("category").apply(lambda x: x.sample(min_samples, random_state=42)).reset_index(drop=True)

# Text preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9’' ]", "", text)
    stopwords = {"va", "bilan", "uchun", "bu", "shuningdek", "haqida"}  # Example stopwords
    text = " ".join([word for word in text.split() if word not in stopwords])
    return text

df_balanced["text"] = df_balanced["text"].apply(preprocess_text)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_balanced["text"], df_balanced["category"], test_size=0.2, random_state=42)

# Create pipeline with TF-IDF and Logistic Regression
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=7000, min_df=1)),
    ("clf", LogisticRegression(max_iter=1000, random_state=42))
])

# Train model
pipeline.fit(X_train, y_train)

def predict_category(text):
    text = preprocess_text(text)
    predicted_category = pipeline.predict([text])[0]
    probabilities = pipeline.predict_proba([text])[0]
    category_probs = dict(zip(pipeline.classes_, probabilities))
    return predicted_category, category_probs

# Example usage
# user_input = "logarifmik funksiya. logarifmik funksiya formulalari"
# predicted, scores = predict_category(user_input)
# print("Predicted Category:", predicted)
#print("Category Probabilities:", scores)
```

```python
user_input = "Amir temurning sulton boyazid ustidan g'alabasi"
predicted, scores = predict_category(user_input)
print("Predicted Category:", predicted)
```
